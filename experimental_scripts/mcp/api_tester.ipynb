```python
# filename: mcp_full_env.py

import os
from dotenv import load_dotenv
import wikipedia
import requests

# ------------------ Load .env ------------------
load_dotenv()
WIKIDATA_USER_AGENT = os.getenv("WIKIDATA_USER_AGENT", "MCPBot/1.0 (contact@example.com)")

# ------------------ MCP RETRIEVAL ------------------
def mcp_retrieve(question):
    contexts = []

    # -------- Wikipedia (Top 3 summaries) --------
    try:
        search_results = wikipedia.search(question, results=3)
        for title in search_results:
            try:
                summary = wikipedia.summary(title, sentences=2)
                if summary.strip():
                    contexts.append(summary.strip())
            except Exception as e:
                # ignore summary fetch errors
                continue
    except Exception as e:
        print("Wikipedia search error:", e)

    # -------- Wikidata (Top 2 descriptions) --------
    try:
        url = "https://www.wikidata.org/w/api.php"
        params = {
            "action": "wbsearchentities",
            "search": question,
            "language": "en",
            "format": "json"
        }
        headers = {"User-Agent": WIKIDATA_USER_AGENT}
        r = requests.get(url, params=params, headers=headers, timeout=5)
        r.raise_for_status()
        data = r.json()
        search = data.get("search", [])
        for item in search[:2]:
            desc = item.get("description","").strip()
            if desc:
                contexts.append(desc)
    except Exception as e:
        print("Wikidata search error:", e)
   

    # -------- Clean empty contexts --------
    contexts = [c for c in contexts if c.strip()]

    # -------- Fallback --------
    if not contexts:
        contexts = ["No context found."]

    return contexts

# ------------------ TEST QUESTIONS ------------------
questions = [
    "What is the tallest mountain in the world?",
    "Who invented the telephone?",
    "What is the speed of light in vacuum?",
    "Which planet is known as the Red Planet?",
    "Who wrote the novel '1984'?",
    "Is the Great Wall of China visible from space?"
]

# ------------------ RUN TEST ------------------
if __name__ == "__main__":
    for q in questions:
        contexts = mcp_retrieve(q)
        print(f"Question: {q}")
        for i, c in enumerate(contexts):
            print(f"  Context {i+1}: {c[:200]}...")  # Show first 200 chars
        print("="*100)
```
